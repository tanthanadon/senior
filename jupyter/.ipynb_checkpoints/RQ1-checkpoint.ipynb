**Calculate MI for each project manually**

```python
n1 = mi_final['Unique Operators']
n2 = mi_final['Unique Operands']
N1 = mi_final['Number of Operators']
N2 = mi_final['Number of Operands']
# Halstead Program Length
N = N1 + N2
N
```

```python
# Halstead Vocabulary
n = n1 + n2
n
```

```python
import numpy as np
# Program Volume
V = N*np.log2(n)
print(V)
mi_final['Code volume'] = V
```

```python
# Program Difficulty
D = (n1/2)*(N2/n2)
print(D)
mi_final['Difficulty'] = D
```

```python
# Program Effort
E = D * V
print(E)
mi_final['Effort'] = E
```

**Normalized factors by logarithm**

```python
# Maintainability Index 
# consisting of SLOC (Source Lines Of Code), Cyclomatic Complexity and Halstead volume.
before = mi_final.copy()
mi_final.mask(mi_final[:] <= 0.0, 1.0, inplace=True)
# Normalzied by log function
project_id = mi_final['project_id']
mi_final = mi_final.apply(np.log)
mi_final['project_id'] = project_id
test = mi_final
test
```

Calculate Maintainability Index following this formula

image.png

```python
HV = test['Effort']
SLOC = test['S Lines of Code']
CC = test['Cyclomatic Complexity']
print(100 * (171-(5.2*np.log(HV))-(0.23*CC)-(16.2*np.log(SLOC)))/171)
MI = np.clip(100 * (171-(5.2*np.log(HV))-(0.23*CC)-(16.2*np.log(SLOC)))/171, 0, 100)
# print(MI)
test['Maintainability Index'] = MI
test
# mi_final.to_csv("mi_final_normalized.csv", index=False)
```

```python
import scipy.stats as scipy
def getPearsonr(x, y):
    return scipy.pearsonr(x,y)        

def getSpearmanr(x, y):
    return scipy.spearmanr(x,y)
```

Caculate correlation and p-values between N and MI for each n-gram

```python
merged_n_mi_original  = mi_final.merge(n_original, how='inner')
# print(merged_n_mi_original)
order = range(1,16)
temp = {}
for i in order:
    x = merged_n_mi_original[merged_n_mi_original['order'] == i][['project_id','cross-entropy']]
    y = merged_n_mi_original[merged_n_mi_original['order'] == i][['project_id','Maintainability Index']]
    a = getPearsonr(x['cross-entropy'], y['Maintainability Index'])[0]
    b = getPearsonr(x['cross-entropy'], y['Maintainability Index'])[1]
    print("##### {0}-gram #####".format(i))
    print("Pearson: correlation = {0}, p-value = {1}".format(a, b))
    # print("Spearman: correlation = {0}, p-value = {1}",a, b)
```
